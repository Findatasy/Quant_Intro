```python
# 动量93年的一篇论文
# http://www.business.unr.edu/faculty/liuc/files/BADM742/Jegadeesh_Titman_1993.pdf

# 一般我们的理解：技术指标--市场是有效的--股价能够迅速反应所有的信息
# 动量的理论基础——市场是不可能完全有效--市场非完全有效（重点在有定“滞后性”，“过激反应”）

import pandas as pd
df= pd.read_csv(r'C:\Users\yons\Desktop\stock_database\00001.csv',encoding = 'GB2312')
df.head()
```

```python
def MOM(df,n,indicator):
    MOM_index = pd.Series(df[indicator].diff(n),name = 'MOM_' +str(n))
    df = df.join(MOM_index)
    return df

df = MOM(df , 30,'open')

print(df)


```
